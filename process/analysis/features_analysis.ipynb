# Imports

```python
import seaborn as sns
import matplotlib.pyplot as plt
import pandas as pd
```

```python
df_grape=pd.read_csv('../../data/GRAPE_QUALITY.csv')
```

```python
df_grape.head()
```

# Harvest day analysis

```python
df_grape['harvest_date'] = pd.to_datetime(df_grape['harvest_date'], errors='coerce')
df_grape['day_of_year'] = df_grape['harvest_date'].dt.dayofyear
df_grape = df_grape.drop(columns=['harvest_date'])
```

```python
plt.figure(figsize=(10, 6))
sns.boxplot(x='quality_category', y='day_of_year', data=df_grape, palette="viridis")
plt.title("Distribution of Harvest Day of the Year by Quality Category")
plt.xlabel("Quality Category")
plt.ylabel("Day of the Year")
plt.show()
```

```python
numerical_features = [
    'sugar_content_brix', 'acidity_ph', 'cluster_weight_g', 'berry_size_mm',
    'day_of_year', 'sun_exposure_hours', 'soil_moisture_percent', 'rainfall_mm'
]

for feature in numerical_features:
    plt.figure(figsize=(10, 6))
    
    sns.violinplot(x='quality_category', y=feature, data=df_grape, palette="viridis", inner="quartile")
    #Boxplot in comment
    #sns.boxplot(x='quality_category', y=feature, data=df_grape, palette="viridis")
    
    plt.title(f"Distribution of {feature} by Quality Category")
    plt.xlabel("Quality Category")
    plt.ylabel(feature)
    plt.show()
```
